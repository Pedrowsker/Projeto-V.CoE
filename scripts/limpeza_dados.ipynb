```python
from glob import glob
import pandas as pd

# 1. Busca os arquivos
arquivos = sorted(glob(r'../data/raw/*.csv'))

lista_df = []

for caminho in arquivos:
    df_tempo = pd.read_csv(caminho)
    
    df_tempo = df_tempo.rename(columns={
        "Order_ID": "ID_Pedido", 
        "Item": "Produto", 
        "Amount": "Quantidade", 
        "Unit_Price": "Preco_Unit", 
        "Date": "Data"
    })
    
    if "int" in caminho:
        df_tempo['Preco_Unit'] = df_tempo['Preco_Unit'] * 5.25
    
    lista_df.append(df_tempo)

df_tab = pd.concat(lista_df, ignore_index=True)


df_tab['Receita_Total'] = df_tab['Quantidade'] * df_tab['Preco_Unit']


display(df_tab)


df_tab.to_excel('../data/processado/base_consolidada.xlsx', index=False)
df_tab.to_csv('../data/processado/base_consolidada.csv', index=False)
```
